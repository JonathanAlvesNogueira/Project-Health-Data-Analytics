```python
import glob
import os
# Caminho para os arquivos (ajuste se necessário)
pasta = r'C:\Users\joth1\Desktop\project_data_helph\Project-Health-Data-Analytics\03_indicador_distancia'

# Padrão para encontrar todos os arquivos desejados
caminhos_arquivos = glob.glob(os.path.join(pasta, 'estimativa_*.csv'))

# Ler e empilhar todos os arquivos
df_total = pd.concat([pd.read_csv(caminho) for caminho in caminhos_arquivos], ignore_index=True)
```
