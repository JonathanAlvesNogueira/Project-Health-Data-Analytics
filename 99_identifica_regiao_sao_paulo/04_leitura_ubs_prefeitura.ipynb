```python
import requests
import tabula
import pandas as pd

# Baixar o PDF
url = "https://www.prefeitura.sp.gov.br/cidade/secretarias/upload/2024_05_08_Relacao_da_UBS_da_Cidade_de_Sao_Paulo%281%29.pdf"
response = requests.get(url)

# Salvar localmente
with open("ubs_sao_paulo.pdf", "wb") as f:
    f.write(response.content)

# Ler o PDF - agora com melhor configuração
tabelas = tabula.read_pdf(
    "ubs_sao_paulo.pdf",
    pages='all',                  # Agora, real mesmo: todas as páginas
    multiple_tables=True,          # Extrair várias tabelas por página
    lattice=True,                  # Melhor para PDFs com linhas separando tabelas
    stream=False
)

# Agora junta tudo num único DataFrame
df_final = pd.concat(tabelas, ignore_index=True)

print(df_final.head())
print(df_final.shape)  # Para ver quantas linhas veio
df_final
```

```python
df_final.to_csv('ubs_sao_paulo.csv', index=False)
```
