```python
import requests
import tabula
import pandas as pd

# Baixar o PDF
url = "https://www.prefeitura.sp.gov.br/cidade/secretarias/upload/saude/arquivos/cis/2024-03-19_Relacao-Estabelecimentos-Completa.pdf"
response = requests.get(url)

# Salvar localmente
with open("hospitais_sp.pdf", "wb") as f:
    f.write(response.content)

# Ler o PDF - agora com melhor configuração
tabelas = tabula.read_pdf(
    "hospitais_sp.pdf",
    pages='all',                  # Agora, real mesmo: todas as páginas
    multiple_tables=True,          # Extrair várias tabelas por página
    lattice=True,                  # Melhor para PDFs com linhas separando tabelas
    stream=False
)

# Agora junta tudo num único DataFrame
df_final = pd.concat(tabelas, ignore_index=True)

print(df_final.head())
print(df_final.shape)  # Para ver quantas linhas veio
df_final
```

```python
df_final.to_csv('hospitais_sao_paulo.csv', index=False)
```
